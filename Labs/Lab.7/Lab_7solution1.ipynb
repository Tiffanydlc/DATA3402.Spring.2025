### First Look

Each row represents a LHC collision event. Each column contains some observable from that event. The variable names are ([based on documentation](https://archive.ics.uci.edu/ml/datasets/SUSY)):

```python
VarNames=["signal", "l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi", "MET", "MET_phi", "MET_rel", "axial_MET", "M_R", "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]
```

Some of these variables represent the "raw" kinematics of the observed final state particles, while others are "features" that are derived from these raw quantities:

```python
RawNames=["l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi", "MET", "MET_phi"]
FeatureNames=list(set(VarNames[1:]).difference(RawNames))
```

We will use pandas to read in the file, and matplotlib to make plots. The following ensures pandas is installed and sets everything up:

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Now we can read the data into a pandas dataframe:

```python
filename = "SUSY.csv"
df = pd.read_csv(filename, dtype='float64', names=VarNames)
```

The first column stores the "truth" label of whether an event was signal or not. Pandas makes it easy to create dataframes that store only the signal or background events:

```python
df_sig=df[df.signal==1]
df_bkg=df[df.signal==0]
```

The following example plots the signal and background distributions of every variable. Note that we use VarNames[1:] to skip the first variable, which was the true label.

```python
import numpy as np
for var in VarNames[1:]:
    print (var)
    plt.figure(figsize=(10,5))
    plt.hist(np.array(df_sig[var]),bins=100,histtype="step", color="red",label="signal",density=1, stacked=True)
    plt.hist(np.array(df_bkg[var]),bins=100,histtype="step", color="blue", label="background",density=1, stacked=True)
    plt.legend(loc='upper right')
    plt.show()
```

## Exercise 3: Make nice figures

Now use `matplotlib` to reproduce as closely as you can figures 5 and 6 from the paper. This exercise is intended to get you to familiarize yourself with making nicely formatted `matplotlib` figures with multiple plots. Note that the plots in the paper are actually wrong!

```python
def create_subplot(ax, variable, xlabel, color_sig='red', color_bkg='blue'):
    ax.hist(df_sig[variable], bins=100, histtype='step',
            color=color_sig, label='Signal', density=True, stacked=True)
    ax.hist(df_bkg[variable], bins=100, histtype='step',
            color=color_bkg, label='Background', density=True, stacked=True)
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel("Arbitrary Units", fontsize=12)


fig5, axes5 = plt.subplots(2, 3, figsize=(15, 8))
fig5.suptitle('Low-level Features', fontsize=16)

# SubPlot: Low level Features
create_subplot(axes5[0, 0], 'l_1_pT', '$p_T(\ell_1)$ (GeV)')
create_subplot(axes5[0, 1], 'l_1_eta', '$\eta(\ell_1)$')
create_subplot(axes5[0, 2], 'l_1_phi', '$\phi(\ell_1)$')
create_subplot(axes5[1, 0], 'l_2_pT', '$p_T(\ell_2)$ (GeV)')
create_subplot(axes5[1, 1], 'l_2_eta', '$\eta(\ell_2)$')
create_subplot(axes5[1, 2], 'l_2_phi', '$\phi(\ell_2)$')
fig5.tight_layout(rect=[0, 0.03, 1, 0.95])
handles, labels = axes5[0, 0].get_legend_handles_labels()
fig5.legend(handles, labels, loc='upper right', fontsize=12)
plt.show()


fig6, axes6 = plt.subplots(2, 3, figsize=(15, 8))
fig6.suptitle('High-level Features', fontsize=16)

# SubPlots: High Level Features
create_subplot(axes6[0, 0], 'MET_rel', '$E^{rel}_T$ miss (GeV)')
create_subplot(axes6[0, 1], 'axial_MET', 'axial $E_T$ miss (GeV)')
create_subplot(axes6[0, 2], 'M_TR_2', '$M_{T2}$ (GeV)')
create_subplot(axes6[1, 0], 'M_R', '$M_R$ (GeV)')
create_subplot(axes6[1, 1], 'S_R', '$S_R$')
create_subplot(axes6[1, 2], 'R', '$R$')
fig6.tight_layout(rect=[0, 0.03, 1, 0.95])
handles, labels = axes6[0, 0].get_legend_handles_labels()
fig6.legend(handles, labels, loc='upper right', fontsize=12)
plt.show()
```

```python
# Part A
from itertools import combinations

def pair_plot(data, columns, sample_size=1000):
    n = len(columns)
    fig, axes = plt.subplots(n, n, figsize=(3*n, 3*n))
    sampled_data = data.sample(sample_size) if sample_size < len(data) else data

    for i in range(n):
        for j in range(n):
            ax = axes[i, j]
            if i == j:
                ax.hist(sampled_data[columns[i]], bins=30, alpha=0.7)
            else:
                ax.scatter(sampled_data[columns[j]], sampled_data[columns[i]], s=1, alpha=0.5)
            if i == n-1:
                ax.set_xlabel(columns[j])
            if j == 0:
                ax.set_ylabel(columns[i])

    plt.tight_layout()
    plt.show()
```

```python
low_level = ['l_1_pT', 'l_1_eta', 'l_1_phi', 'MET']
high_level = ['M_Delta_R', 'M_R', 'MT2']
pair_plot(df, low_level)
pair_plot(df, high_level)
```

```python
#Part B
def fast_pair_plot(data, columns, sample_size=1000):
    sampled_data = data[columns].sample(sample_size).to_numpy()
    n = len(columns)
    fig, axes = plt.subplots(n, n, figsize=(3*n, 3*n))

    for i in range(n):
        for j in range(n):
            ax = axes[i, j]
            if i == j:
                ax.hist(sampled_data[:, i], bins=30, alpha=0.7)
            else:
                ax.plot(sampled_data[:, j], sampled_data[:, i], 'o', markersize=1, alpha=0.5)
            if i == n-1:
                ax.set_xlabel(columns[j])
            if j == 0:
                ax.set_ylabel(columns[i])
    plt.tight_layout()
    plt.show()
```

```python
fast_pair_plot(df, low_level)
fast_pair_plot(df, high_level)
```
